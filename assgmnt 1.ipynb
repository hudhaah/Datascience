```python
"Arithmetic operations"

n=int(input("enter the first value :"))
m=int(input("enter the second value :"))
while(1):
    print("1.SUM\n2.DIFFERENCE\n3.MULTIPLY\n4.DIVIDE\n5.EXIT")
    ch=int(input("Enter your choice:"))
    if(ch==1):
        print("Sum=",n+m)
    elif(ch==2):
        print("Difference=",n-m)
    elif(ch==3):
        print("Product=",n*m)
    elif(ch==4):
        print("Quotient=",n/m)
    else:
        print("EXITING")
        break
```

```python
"String operations"

s1=str(input("Enter the first string:"))
s2=str(input("Enter the second strig:"))
while(1):
    print("1.Length of string\n2.Slicing of string\n3.To uppercase\n4.To lowercase\n5.Replace\n6.Concatenate\n7.Exit")
    ch=int(input("Enter your choice:"))
    if(ch==1):
        print("Length of string 1=",len(s1))
    elif(ch==2):
        print(s1[2:])
    elif(ch==3):
        print("Uppercase=",s1.upper())
    elif(ch==4):
        print("Lowercase=",s1.lower())
    elif(ch==5):
        print("After replacing=",s2.replace("o","r"))
    elif(ch==6):
        s=s1+s2
        print("Concatenation of string=",s)
    else:
        print("Exiting")
        break
```
